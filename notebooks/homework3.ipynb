```python
#problem 2:
import networkx as nx
import matplotlib.pyplot as plt

#G = nx.complete_graph(10)
#G = nx.barbell_graph(6, 3)
#G = nx.cycle_graph(12)
#G = nx.circulant_graph(10, [1, 3])
#G = nx.grid_2d_graph(5, 5)
G = nx.ladder_graph(10)

plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color="lightblue", node_size=800)
plt.show()
```

```python
#problem 3
G = nx.barabasi_albert_graph(25, 2)

layouts = {
    "Spring Layout": nx.spring_layout(G),
    "Circular Layout": nx.circular_layout(G),
    "Neato Layout": nx.nx_pydot.graphviz_layout(G, prog="neato"),
}

plt.figure(figsize=(15, 5))

for i, (name, pos) in enumerate(layouts.items(), 1):
    plt.subplot(1, 3, i)
    nx.draw(G, pos, with_labels=True, node_size=500)
    plt.title(name)

plt.show()
```

```python
#problem 4
import numpy as np

G = nx.barbell_graph(8, 4)
pos = nx.spring_layout(G)

n = G.number_of_nodes()

node_colors = np.linspace(0, 1, n)
node_sizes = np.linspace(300, 1400, n)

plt.figure(figsize=(7, 7))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=node_colors,
    node_size=node_sizes,
    cmap=plt.cm.plasma,   
    font_size=10
)

plt.title("Barbell Graph with Unique Colors and Sizes")
plt.show()

```
